```
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from numpy import mean
```

```
#Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials

results = []
for num_trials in range(1,1000):
    trials = np.random.randint(1, 7, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})

df.plot()
plt.title("Normal distribution")
```

```
results = []
for num_trials in range(1,1000):
    trials = np.random.gamma(1, 7, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
plt.title("Gamma distribution")
```

```

#Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

means = [mean(np.random.gamma(1, 7, 1000)) for _ in range(1000)]


plt.hist(means)
plt.title('Central Limit Thm for Gamma')
plt.show()
```
