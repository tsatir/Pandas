```
#Find all successful documentary projects and sort them by the amount pledged. Print the top 10 highest pledges.
ksprojects1 = ksprojects.iloc[((ksprojects['state']=='successful') & (ksprojects['category'] == 'Documentary')).values]
ksprojects1.sort_values(by='pledged', ascending=False).head(10)
```

```
#Create a new column named average_per_backer and set its value to the total amount pledged / number of backers. 
ksprojects2 = ksprojects.assign(average_per_backer = ksprojects.pledged / ksprojects.backers )
#What happened to the rows with 0 backers? 
ksprojects2.iloc[(ksprojects2['backers']==0).values].head(3)
```

We are trying to divide pledged to "0", it couldn't calculate and results with a Nan value. We can drop Nan values or change the nan values to "No backer"

```
#How can this be dealt with?
ksprojects3 = ksprojects2.replace(np.nan, 'No backer')
ksprojects3.head()
```

```
#Drop all rows with 0 backers then repeat the previous exercise.
ksprojects4 = ksprojects[ksprojects.backers != 0]
#Create a new column named average_per_backer and set its value to the total amount pledged / number of backers. 
ksprojects5 = ksprojects4.assign(average_per_backer = ksprojects4.pledged / ksprojects4.backers )
ksprojects5.head()
```

```
###census
census_population = pd.read_csv("https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv")
census_regions = pd.read_csv("https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv")
```

```
census_population.info()
```

```
census_regions.info()
```

```
#Merge these two DataFrames into a single DataFrame called census
census = pd.merge(census_regions,census_population)
census.info()
print(any(census.duplicated()))
```

```

census.head()
```

```
#Re-shape census so that one column contains all population measures, and another the year attributes.
census.head(3)
reshape_census = pd.melt(frame=census, id_vars=['state','region','division'], var_name='year_id', value_name='population')
reshape_census.head(3)

#census_melt = pd.melt(census, id_vars=["state", "region", "division"], var_name="year", value_name="population")
#census_melt.head()
```

```
#Group the data by year and summarize it.
reshape_census.groupby(['year_id']).describe()
```

```
#Group the data by region, division and year and summarize it.
reshape_census.groupby(['region','division','year_id']).describe()
```
